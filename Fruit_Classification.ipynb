```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import cross_val_score,GridSearchCV
```

```python
fruits=pd.read_csv('fruits_data.csv')
fruits.info()
```

```python
fruits.head()
```

```python
encoder=LabelEncoder()
fruits['label']=encoder.fit_transform(fruits['name'])
fruits.head()
```

```python
x=fruits[['mass','width','height','cscore']]
y=fruits['name']
```

```python
decision_tree=DecisionTreeClassifier()
decision_tree_scores=cross_val_score(decision_tree,x,y,cv=5)
decision_tree_scores
```

```python
classifier=tree.DecisionTreeClassifier()
classifier=classifier.fit(x,y)
classifier.predict([[10,4,3,38]])
```

```python
np.mean(decision_tree_scores)
```

```python
adaboost=AdaBoostClassifier(estimator=decision_tree,n_estimators=100)
adaboost_scores=cross_val_score(adaboost,x,y,cv=5)
adaboost_scores
```

```python
classifier=tree.DecisionTreeClassifier()
classifier=classifier.fit(x,y)
classifier.predict([[10,4,3,38]])
```
